# Resizing the dataset and loading

```python
import cv2
h, w = 50, 250
```

```python
def dataset():
    data = []
    path = r'C:\Users\ADMIN\remaining'
    for img in os.listdir(path):
        img_path = os.path.join(path, img)
        img_array = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
        img_array = img_array.astype(np.float32)
        new_array = resize(img_array, (h, w))
        label = os.path.basename(img_path).split('.png')[0]
        data.append([new_array, label])
    return data
```

```python
data = dataset()
```

```python
len(data)
```

```python
X = []
y = []
```
